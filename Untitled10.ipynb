```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
plt.style.use('seaborn-v0_8')

# Chargement
df = pd.read_csv(r'C:\Users\yanis\Downloads\archive (1)\global_ads_performance_dataset.csv')

# Premier aperçu
print(df.shape)
print(df.head())
print(df.info())
print(df.columns.tolist())
```

```python
# Résumé global
print("=== RÉSUMÉ GLOBAL ===")
print(f"Période : {df['date'].min()} au {df['date'].max()}")
print(f"Plateformes : {df['platform'].unique()}")
print(f"Industries : {df['industry'].unique()}")
print(f"Pays : {df['country'].unique()}")
print(f"Budget total dépensé : {df['ad_spend'].sum():,.0f} €")
print(f"Revenue total généré : {df['revenue'].sum():,.0f} €")
print(f"ROAS global : {(df['revenue'].sum() / df['ad_spend'].sum()):.2f}")

# Comparaison des 3 plateformes
print("\n=== PERFORMANCE PAR PLATEFORME ===")
plateforme = df.groupby('platform').agg(
    budget=('ad_spend', 'sum'),
    revenue=('revenue', 'sum'),
    conversions=('conversions', 'sum'),
    CTR_moyen=('CTR', 'mean'),
    CPC_moyen=('CPC', 'mean'),
    CPA_moyen=('CPA', 'mean'),
    ROAS_moyen=('ROAS', 'mean')
).round(2)

plateforme['budget_%'] = (plateforme['budget'] / plateforme['budget'].sum() * 100).round(1)
plateforme['ROAS_reel'] = (plateforme['revenue'] / plateforme['budget']).round(2)
print(plateforme.sort_values('ROAS_reel', ascending=False))

# Visualisation
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

plateformes = plateforme.index.tolist()
colors = ['#4285F4', '#1877F2', '#010101']  # Google bleu, Meta bleu, TikTok noir

# ROAS réel
axes[0,0].bar(plateformes, plateforme['ROAS_reel'], color=colors)
axes[0,0].set_title('ROAS réel par plateforme (plus haut = mieux)')
axes[0,0].set_xlabel('Plateforme')
axes[0,0].axhline(y=plateforme['ROAS_reel'].mean(), color='red', linestyle='--', label=f'Moyenne : {plateforme["ROAS_reel"].mean():.2f}')
axes[0,0].legend()

# CPA moyen
axes[0,1].bar(plateformes, plateforme['CPA_moyen'], color=colors)
axes[0,1].set_title('CPA moyen par plateforme (plus bas = mieux)')
axes[0,1].set_xlabel('Plateforme')
axes[0,1].axhline(y=plateforme['CPA_moyen'].mean(), color='red', linestyle='--', label=f'Moyenne : {plateforme["CPA_moyen"].mean():.2f}')
axes[0,1].legend()

# CTR moyen
axes[1,0].bar(plateformes, plateforme['CTR_moyen'] * 100, color=colors)
axes[1,0].set_title('CTR moyen par plateforme (%)')
axes[1,0].set_xlabel('Plateforme')

# Budget vs Revenue
x = range(len(plateformes))
width = 0.35
axes[1,1].bar([i - width/2 for i in x], plateforme['budget'], width, label='Budget dépensé', color='#E8593C', alpha=0.8)
axes[1,1].bar([i + width/2 for i in x], plateforme['revenue'], width, label='Revenue généré', color='#1D9E75', alpha=0.8)
axes[1,1].set_xticks(list(x))
axes[1,1].set_xticklabels(plateformes)
axes[1,1].set_title('Budget dépensé vs Revenue généré (€)')
axes[1,1].legend()

plt.tight_layout()
plt.savefig('analyse_plateformes.png', dpi=150, bbox_inches='tight')
plt.show()
print("\nGraphique sauvegardé !")
```

```python
# Analyse par type de campagne
print("=== PERFORMANCE PAR TYPE DE CAMPAGNE ===")
camp_type = df.groupby(['platform', 'campaign_type']).agg(
    budget=('ad_spend', 'sum'),
    ROAS_reel=('ROAS', 'mean'),
    CPA=('CPA', 'mean'),
    conversions=('conversions', 'sum')
).round(2)
print(camp_type.sort_values('ROAS_reel', ascending=False).head(10))

# Analyse par industrie
print("\n=== TOP INDUSTRIES PAR ROAS ===")
industrie = df.groupby('industry').agg(
    budget=('ad_spend', 'sum'),
    revenue=('revenue', 'sum'),
    CPA_moyen=('CPA', 'mean'),
    conversions=('conversions', 'sum')
).round(2)
industrie['ROAS_reel'] = (industrie['revenue'] / industrie['budget']).round(2)
print(industrie.sort_values('ROAS_reel', ascending=False))

# Visualisation
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

camp_pivot = df.groupby(['platform', 'campaign_type'])['ROAS'].mean().unstack()
camp_pivot.plot(kind='bar', ax=axes[0], colormap='viridis')
axes[0].set_title('ROAS moyen par plateforme et type de campagne')
axes[0].set_xlabel('Plateforme')
axes[0].tick_params(axis='x', rotation=15)
axes[0].legend(title='Type campagne', fontsize=9)

industrie_sorted = industrie.sort_values('ROAS_reel', ascending=True)
axes[1].barh(industrie_sorted.index, industrie_sorted['ROAS_reel'], color='#7F77DD')
axes[1].set_title('ROAS réel par industrie')
axes[1].set_xlabel('ROAS')
axes[1].axvline(x=industrie_sorted['ROAS_reel'].mean(), color='red',
                linestyle='--', label=f'Moyenne : {industrie_sorted["ROAS_reel"].mean():.2f}')
axes[1].legend()

plt.tight_layout()
plt.savefig('analyse_campagnes_industries.png', dpi=150, bbox_inches='tight')
plt.show()
print("Graphique sauvegardé !")
```

```python
print("=== ANALYSE FINALE & RECOMMANDATIONS ===")
print("""
TYPE DE CAMPAGNE — Les gagnants clairs
- TikTok Search : ROAS 10.52 — champion absolu
- TikTok Display : ROAS 10.04 — très solide
- Google Ads : ROAS 3.47-4.79 — le moins rentable malgré 57% du budget
→ RECOMMANDATION : Basculer une partie du budget Google vers TikTok Search

INDUSTRIE — Le marché le plus rentable
- SaaS : ROAS 5.04 — meilleur secteur
- E-commerce : ROAS 5.01 — excellent pour notre niche cible
- Fintech : ROAS 4.48 — le moins performant
→ RECOMMANDATION : Prioriser SaaS et E-commerce pour les campagnes
""")

# Analyse tendance mensuelle
df['date'] = pd.to_datetime(df['date'])
df['mois'] = df['date'].dt.to_period('M')

tendance = df.groupby(['mois', 'platform']).agg(
    ROAS=('ROAS', 'mean'),
    budget=('ad_spend', 'sum')
).round(2).reset_index()

tendance['mois_str'] = tendance['mois'].astype(str)

# Visualisation tendance + recommandations budget
fig, axes = plt.subplots(2, 1, figsize=(16, 12))

# Évolution ROAS mensuel par plateforme
colors_plt = {'Google Ads': '#4285F4', 'TikTok Ads': '#010101', 'Meta Ads': '#1877F2'}
for platform in df['platform'].unique():
    data = tendance[tendance['platform'] == platform]
    axes[0].plot(data['mois_str'], data['ROAS'], 
                marker='o', label=platform, 
                color=colors_plt[platform], linewidth=2)

axes[0].set_title('Évolution du ROAS mensuel par plateforme (2024)')
axes[0].set_xlabel('Mois')
axes[0].set_ylabel('ROAS moyen')
axes[0].legend()
axes[0].tick_params(axis='x', rotation=45)

# Simulation réallocation budget
labels = ['Situation actuelle', 'Budget optimisé']
google = [57.2, 30]
meta = [19.0, 25]
tiktok = [23.9, 45]

x = range(len(labels))
width = 0.25
axes[1].bar([i - width for i in x], google, width, label='Google Ads', color='#4285F4', alpha=0.8)
axes[1].bar([i for i in x], meta, width, label='Meta Ads', color='#1877F2', alpha=0.8)
axes[1].bar([i + width for i in x], tiktok, width, label='TikTok Ads', color='#555555', alpha=0.8)
axes[1].set_title('Réallocation budgétaire recommandée (%)')
axes[1].set_ylabel('% du budget')
axes[1].set_xticks(list(x))
axes[1].set_xticklabels(labels)
axes[1].legend()

# Calcul impact réallocation
budget_total = 11108749
roas_actuel = 4.88
roas_optimise = (0.30 * 3.47) + (0.25 * 5.66) + (0.45 * 7.62)
revenue_actuel = budget_total * roas_actuel
revenue_optimise = budget_total * roas_optimise

print(f"=== IMPACT RÉALLOCATION BUDGÉTAIRE ===")
print(f"ROAS actuel : {roas_actuel:.2f}")
print(f"ROAS optimisé estimé : {roas_optimise:.2f}")
print(f"Revenue actuel : {revenue_actuel:,.0f} €")
print(f"Revenue optimisé : {revenue_optimise:,.0f} €")
print(f"Gain potentiel : +{revenue_optimise - revenue_actuel:,.0f} € (+{((revenue_optimise/revenue_actuel)-1)*100:.0f}%)")

plt.tight_layout()
plt.savefig('analyse_finale_multiplateforme.png', dpi=150, bbox_inches='tight')
plt.show()
print("\nAnalyse complète terminée !")
```
